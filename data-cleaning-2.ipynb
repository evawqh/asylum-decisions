```python
import pandas as pd
```

```python
df_af = pd.read_csv("asylum_affirmative.csv")
df_def = pd.read_csv("asylum_defensive.csv")
df_ref = pd.read_csv("refugees.csv")
```

```python
# Replace not disclosed "D" with -1 and not applicable "X" with -2.
# "-" means 0, according to the documentation
df_af = df_af.replace({"D": -1, "-": 0, "X": -2})
```

```python
df_af["2021"] = df_af["2021"].astype("float")
df_af["2020"] = df_af["2020"].astype("float")
df_af["2019"] = df_af["2019"].astype("float")
df_af["2018"] = df_af["2018"].astype("float")
df_af["2017"] = df_af["2017"].astype("float")
df_af["2016"] = df_af["2016"].astype("float")
df_af["2015"] = df_af["2015"].astype("float")
df_af["2014"] = df_af["2014"].astype("float")
df_af["2013"] = df_af["2013"].astype("float")
df_af["2012"] = df_af["2012"].astype("float")
```

```python
df_def = df_def.replace({"D": -1, "-": 0, "X": -2, "NaN": -2})
```

```python
df_def["2021"] = df_def["2021"].astype(float)
df_def["2020"] = df_def["2020"].astype(float)
df_def["2019"] = df_def["2019"].astype(float)
df_def["2018"] = df_def["2018"].astype(float)
df_def["2017"] = df_def["2017"].astype(float)
df_def["2016"] = df_def["2016"].astype(float)
df_def["2015"] = df_def["2015"].astype(float)
df_def["2014"] = df_def["2014"].astype(float)
df_def["2013"] = df_def["2013"].astype(float)
df_def["2012"] = df_def["2012"].astype(float)
```

```python
df_ref = df_ref.replace({"D": -1, "-": 0, "X": -1})
```

```python
df_ref["2021"] = df_ref["2021"].astype(float)
df_ref["2020"] = df_ref["2020"].astype(float)
df_ref["2019"] = df_ref["2019"].astype(float)
df_ref["2018"] = df_ref["2018"].astype(float)
df_ref["2017"] = df_ref["2017"].astype(float)
df_ref["2016"] = df_ref["2016"].astype(float)
df_ref["2015"] = df_ref["2015"].astype(float)
df_ref["2014"] = df_ref["2014"].astype(float)
df_ref["2013"] = df_ref["2013"].astype(float)
df_ref["2012"] = df_ref["2012"].astype(float)
```

```python
df_af = df_af.sort_values(by="2021", ascending=False, na_position="last")
df_af.head(21)
```

```python
df_def = df_def.sort_values(by="2021", ascending=False, na_position="last")
df_def.head(21)
```

```python
df_ref = df_ref.sort_values(by="2021", ascending=False, na_position="last")
df_ref.head(21)
```
